## 16.4
Use the `sqlite3` module to create a SQLite database called `books.db` and a table called `books` with these fields: `title` (text), `author` (text), and `year` (integer).

````python
import sqlite3

# Connecting to db and setting up the table fields
# Note: I'm using 'book' as the table name since later prompts use the singular vers.
db = sqlite3.connect('books.db')
cursor = db.cursor()
cursor.execute('CREATE TABLE IF NOT EXISTS book (title text, author text, year int)')
db.commit()
````

## 16.5
Read `books2.csv` and insert its data into the `book` table.

````python
import csv

# Reading csv I made earlier and moving the data into the sqlite table
with open('books2.csv', 'r') as infile:
    books = csv.DictReader(infile)
    for row in books:
        cursor.execute('INSERT INTO book VALUES (?, ?, ?)', (row['title'], row['author'], row['year']))
db.commit()
````

## 16.8
Use the `sqlalchemy` module to connect to the sqlite3 database `books.db` that you just made in exercise 16.4. As in 16.6, select and print the `title` column from the `book` table in alphabetical order.

````python
import sqlalchemy

# Doing same title search but using sqlalchemy this time
engine = sqlalchemy.create_engine('sqlite:///books.db')
with engine.connect() as conn:
    query = sqlalchemy.text('SELECT title FROM book ORDER BY title ASC')
    result = conn.execute(query)
    for row in result:
        print(row[0])

# Closing original sqlite connection just to be clean
db.close()
````
